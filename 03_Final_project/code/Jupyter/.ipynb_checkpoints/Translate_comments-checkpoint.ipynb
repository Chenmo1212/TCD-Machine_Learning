```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from googletrans import Translator
import googletrans
import time
from langdetect import detect, LangDetectException
```

```python
df = pd.read_csv('reviews.csv')
```

```python
comments = df['comments']

# Handle null values
comments = comments.fillna('')
comments = comments.str.replace('<br/>','')
```

```python
def is_punctuation(str_input):
    return bool(re.match('^[\u0000-\u002f\u003a-\u0040\u005b-\u0060\u007b-\u007f]+$', str_input))

def is_all_num_and_punc(s):
    return bool(re.match(r'^[\d\W]+$',s))
```

```python
comments.head
```

```python
test_comments = comments[:10]
```

```python
translator = Translator()

count = 0
cache = {}
```

```python
def translate(x, idx, count, cache):
    try:
        if is_punctuation(x):
            print('is_punctuation', idx, ':', x)
            cache[idx] = x
            return x
        
        if is_all_num_and_punc(x) and x != '':
            print('is_all_num_and_punc', idx , ':', x)
            cache[idx] = x
            return x

        if detect(x) != "en":
            count = count + 1
    except LangDetectException:
        print(idx, 'LangDetectException')
    except Exception as e:
        print(idx, e)


for idx in comments.index:
#     print(idx)
    comments[idx] = translate(comments[idx], idx, count, cache)

count
```

```python
import csv

with open('cache.txt', 'w', encoding='utf-8') as f:
    for key, value in cache.items():
        f.write('%s,%s\n' % (key, value))
        

with open('cache.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    for key, value in cache.items():
        writer.writerow([key, value])
```
