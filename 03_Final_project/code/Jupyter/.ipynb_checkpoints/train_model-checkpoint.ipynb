```python
def plotErrorBar(x, mean, var, xlabel, title, image):
    # create an index for each tick position
    xi = list(range(len(x)))
    npc = np.array(xi)

    fig, ax = plt.subplots()
    plt.errorbar(npc, mean, var, linewidth=3, marker='o', c='r')

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Mean')
    plt.xticks(xi, x)
    fig.show()
```

```python
def MAE(name, model, Xtrain, Xtest, ytrain, ytest):
    ypred = model.predict(Xtrain)
    mae_train = mean_absolute_error(ytrain, ypred)
    ypred = model.predict(Xtest)
    mae_test = mean_absolute_error(ytest, ypred)
    print("MAE    Train:%f      Test:%f"%(mae_train, mae_test))
    #print("The MAE for the %s model with the training data was %f and with the test data was %f"%(name, mae_train, mae_test))

def MSE(name, model, Xtrain, Xtest, ytrain, ytest):
    ypred = model.predict(Xtrain)
    mse_train = mean_squared_error(ytrain, ypred, squared=True)
    ypred = model.predict(Xtest)
    mse_test = mean_squared_error(ytest, ypred, squared=True)
    print("MSE    Train:%f      Test:%f" % (mse_train, mse_test))
    #print("The RMSE for the %s model with the training data was %f and with the test data was %f" % (name, mse_train, mse_test))

def r_sq(name, model, Xtrain, Xtest, ytrain, ytest):
    ypred = model.predict(Xtrain)
    rsq_train = r2_score(ytrain, ypred)
    ypred = model.predict(Xtest)
    rsq_test = r2_score(ytest, ypred)
    print("RSQ    Train:%f      Test:%f" % (rsq_train, rsq_test))
    #print("The R squared score for the %s model with the training data was %f and with the test data was %f" % (name, rsq_train, rsq_test))

def plot(name, model, Xtrain, Xtest, ytrain, ytest, filename):
    ypred = model.predict(Xtrain)
    fig, axs = plt.subplots()
    axs.scatter(Xtrain[:,1], ytrain, s=20, c='b', marker='+')
    axs.scatter(Xtrain[:,1], ypred, s=20, c='r', marker='+')
    axs.set_title(name + ' - Precicted Training Data')
    axs.set_xlabel('feature')
    axs.set_ylabel('amount')
    axs.legend(['Target', 'Predictions'])
    fig.show()
    #fig.savefig('Plots/Prediction Plots/' + filename + '_train_pred')

    ypred = model.predict(Xtest)
    fig, axs = plt.subplots()
    axs.scatter(Xtest[:, 1], ytest, s=20, c='b', marker='+')
    axs.scatter(Xtest[:, 1], ypred, s=20, c='r', marker='+')
    axs.set_title(name + ' - Predicted Test Data')
    axs.set_xlabel('feature')
    axs.set_ylabel('amount')
    axs.legend(['Target', 'Predictions'])
    fig.show()
    
def plot_model_res(model, name):
    strategy = "mean"
    print("{}:".format(name))
    MAE("{}".format(name), model, X_train, X_test, y_train, y_test)
    MSE("{}".format(name), model, X_train, X_test, y_train, y_test)
    r_sq("{}".format(name), model, X_train, X_test, y_train, y_test)
#     plot("{} Regression".format(name), model, X_train, X_test, y_train, y_test, "d")

def normalization(data):
    normalized_df = pd.DataFrame()

    # 对每一列进行正则化
    for column in data.columns:
        normalized_df[column] = data[column] / data[column].max()
    
    return normalized_df
```

```python
# Define a function to find the optimal hyperparameters
def find_optimal_hyperparameters(model, hyperparameters, title):
    # Initialize list for MSEs
    mse_list = []

    # Loop over the hyperparameters
    for hyperparameter in hyperparameters:
        # Set the hyperparameter
        model.set_params(**hyperparameter)
        
        # Perform cross-validation
        from sklearn.model_selection import cross_val_score
        scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
        
        # Compute mean MSE
        mse = -scores.mean()
        
        # Append to list
        mse_list.append(mse)

    hyp_list = [hyperparameter[key] for hyperparameter in hyperparameters for key in hyperparameter]
    plt.plot(hyp_list, mse_list)
    plt.xlabel('Hyperparameters value')
    plt.ylabel('MSE')
    plt.title(title)
    plt.show()
    
    # Find the index of the minimum mse
    min_idx = mse_list.index(min(mse_list))
    
    # Output the hyperparameter corresponding to the minimum MSE
    return hyperparameters[min_idx], min(mse_list)
```

## SVM

```python
# Define the model
svr_model = SVR()

# Define the range of C parameters
svr_hypers = [{'C': 0.001}, {'C': 0.01}, {'C': 0.1}, {'C': 1}, {'C': 10}, {'C': 100}, {'C': 200}]

# Get the optimal hyperparameter
svr_hyper, svr_mse = find_optimal_hyperparameters(svr_model, svr_hypers, 'Correspondence of SVM hyperparticipation in MSE')
print('The optimal hyperparameters are:', svr_hyper)
print('The minimum MSE are:', svr_mse)
```

```python
SVM_list = []
KNN_list = []
Lasso_list = []
Ridge_list = []
Dummy_list = []


for tar in features.keys():
    feature_list = features[tar]
    target = targets[tar]
    
    df = data[feature_list]
    
    # 对每一列进行正则化
    for col in df.columns:
        norm(df[col])

    # Separate features and target
    X = df[feature_list]
    y = data[target]
    norm(y)
    
    # Split train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    
    # svm
    svr_model = SVR()
    svr_hypers = [{'C': 0.001}, {'C': 0.01}, {'C': 0.1}, {'C': 1}, {'C': 10}, {'C': 100}, {'C': 200}]
    svr_hyper, svr_mse = find_optimal_hyperparameters(svr_model, svr_hypers, 'Correspondence of SVM hyperparticipation in MSE')
    SVM_list.append(svr_mse)
    
    # knn
    knn_model = KNeighborsRegressor()
    knn_hypers = [{'n_neighbors': 1}, {'n_neighbors': 5}, {'n_neighbors': 10}, {'n_neighbors': 15}, 
                  {'n_neighbors': 20}, {'n_neighbors': 25}, {'n_neighbors': 30}, {'n_neighbors': 50}]
    knn_hyper, knn_mse = find_optimal_hyperparameters(knn_model, knn_hypers, 'Correspondence of Knn hyperparticipation in KNN')
    KNN_list.append(knn_mse)
    
    # lasso
    lasso_model = Lasso()
    lasso_hypers = [{'alpha': 0.000001}, {'alpha': 0.00001}, {'alpha': 0.0001}, {'alpha': 0.001}]
    lasso_hyper, lasso_mse = find_optimal_hyperparameters(lasso_model, lasso_hypers, 'Correspondence of Lasso hyperparticipation in Lasso')
    Lasso_list.append(lasso_mse)
    
    # ridge
    Ridge_model = Ridge()
    Ridge_hypers = [{'alpha': 0.001}, {'alpha': 0.01}, {'alpha': 0.1}, {'alpha': 1}, {'alpha': 10}, {'alpha': 100}]
    Ridge_mse_list, ridge_mse = find_optimal_hyperparameters(Ridge_model, Ridge_hypers, 'Correspondence of Ridge hyperparticipation in Ridge')
    Ridge_list.append(ridge_mse)
    
    # Create the dummy model
    dummy_model = DummyRegressor()
    dummy_model.fit(X_train, y_train)
    y_pred = dummy_model.predict(X_test)
    dummy_mse = mean_squared_error(y_test, y_pred)
    Dummy_list.append(dummy_mse)
```
